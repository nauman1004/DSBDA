```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

csv_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
col_names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Species']
iris = pd.read_csv(csv_url, header=None, names=col_names)

print(iris.head(30))

# Selecting only the numeric columns
numeric_iris = iris.select_dtypes(include=[np.number])

# Calculating the range
range_values = numeric_iris.max() - numeric_iris.min()

# Calculating the variance
variance_values = numeric_iris.var()

# Calculating the standard deviation
std_dev_values = numeric_iris.std()

# Calculating the interquartile range (IQR)
IQR_values = numeric_iris.quantile(0.75) - numeric_iris.quantile(0.25)

# Finding the minimum and maximum values for each column
min_values = numeric_iris.min()
max_values = numeric_iris.max()

# Displaying the results
print("Range values:\n", range_values)
print("\nVariance values:\n", variance_values)
print("\nStandard Deviation values:\n", std_dev_values)
print("\nInterquartile Range (IQR) values:\n", IQR_values)
print("\nMinimum values:\n", min_values)
print("\nMaximum values:\n", max_values)

# Visualization

# 1. Histograms for each feature
numeric_iris.hist(bins=20, figsize=(12, 8))
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.show()

# 2. Boxplots for each feature
plt.figure(figsize=(12, 8))
numeric_iris.plot(kind='box', vert=False, patch_artist=True)
plt.title('Boxplots of Numeric Features')
plt.show()


```
